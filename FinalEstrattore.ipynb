```python
import nltk
import csv
import string
import re
import json
import pandas as pd
from nltk.corpus import wordnet
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import WordPunctTokenizer

punt=WordPunctTokenizer()
wordnet_lemmatizer = WordNetLemmatizer()

in_file = open(r"C:\Users\franc\Desktop\bando1.txt","r")
bando1 = in_file.read()
in_file2 = open(r"C:\Users\franc\Desktop\bando2.txt","r")
bando2 = in_file2.read()

bandi=[bando1,bando2]

def stampa_controllo(array):
    n=2
    for word in array:
        print(word)
        print ("\n")
        print("---------------------------------------------------------------------------------------------------------------------------------------")
        print("Output bando "+str(n)+": ")
        n=n+1
        
def EstrattoreNUTS(text): 
    regex="\\bit(c(1[1-8]{0,1}|20{0,1}|3[1-4]{0,1}|4((?!5)[1-9]|[A-D]){0,1}){0,1}|(h(10{0,1}|20{0,1}|3[1-7]{0,1}|4[1-4]{0,1}|5[1-9]{0,1}){0,1})|i(1([1-9]|A){0,1}|2[1-2]{0,1}|3[1-5]{0,1}|4[1-5]{0,1}){0,1}|f(1[1-4]{0,1}|2[1-2]{0,1}|3[1-5]{0,1}|4[3-8]{0,1}|5[1-2]{0,1}|6[1-5]{0,1}){0,1}|g(1[1-9]{0,1}|2([5-9]|[A-C]){0,1}){0,1})\\b"
    i_nuts=[]
    for w in text:
        if "nuts" in w:
            match=list(re.finditer(regex,w))
            for trovato in match:
                nuts=trovato.group()
                i_nuts.append(nuts)
    return i_nuts
  
    
def EstrattoreDateScadenza(text):
    date=[]
    regex ="((((((0[1-9]|1[0-9]|2[0-8])[\\ | \\- | \\/ | \\.](0[1-9]|1[012]))|((29|30|31)[\\ | \\- | \\/ | \\.](0[13578]|1[02]))|((29|30)[\\ | \\- | \\/ | \\.](0[4,6,9]|11)))[\\ | \\- | \\/ | \\.](19|[2-9][0-9])\\d\\d)|(29[\\ | \\- | \\/ | \\.]02[\\ | \\- | \\/ | \\.](19|[2-9][0-9])(00|04|08|12|16|20|24|28|32|36|40|44|48|52|56|60|64|68|72|76|80|84|88|92|96)|(gennaio|febbraio|marzo|aprile|maggio|giugno|luglio|agosto|settembre|ottobre|novembre|dicembre)\\ (19|[2-9][0-9][0-9][0-9])|(\\ ([0-2]*[0-9]|3[0-1])\\ (gennaio|febbraio|marzo|aprile|maggio|giugno|luglio|agosto|settembre|ottobre|novembre|dicembre)\\ (19|[2-9][0-9][0-9][0-9]))|(\\ ([0-2]*[0-9]|3[0-1])\\ (gennaio|febbraio|marzo|aprile|maggio|giugno|luglio|agosto|settembre|ottobre|novembre|dicembre)))))"
    for w in text:
         if "ultimazione" in w:
            date_chiusura="ultimazione (...)*"
            app=re.search(date_chiusura,w)
            if app:
                match=re.search(regex,app.group())
                if match:
                    data=match.group()
                    date.append(data)
         if "scadenza" in w:
            date_chiusura="scadenza (...)*"
            app=re.search(date_chiusura,w)
            if app:
                match=re.search(regex,app.group())
                if match:
                    data=match.group()
                    date.append(data)
         if "termine" in w:
            date_chiusura="termine (...)*(...)*"
            app=re.search(date_chiusura,w)
            if app:
                match=re.search(regex,app.group())
                if match:
                    data=match.group()
                    date.append(data)
         if "entro le ore"in w:
            date_chiusura="entro le ore (...)*(...)*"
            app=re.search(date_chiusura,w)
            if app:
                match=re.search(regex,app.group())
                if match:
                    data=match.group()
                    date.append(data)
    return date
    
def EstrattoreDateApertura(text):
    date=[]
    regex ="((((((0[1-9]|1[0-9]|2[0-8])[\\ | \\- | \\/ | \\.](0[1-9]|1[012]))|((29|30|31)[\\ | \\- | \\/ | \\.](0[13578]|1[02]))|((29|30)[\\ | \\- | \\/ | \\.](0[4,6,9]|11)))[\\ | \\- | \\/ | \\.](19|[2-9][0-9])\\d\\d)|(29[\\ | \\- | \\/ | \\.]02[\\ | \\- | \\/ | \\.](19|[2-9][0-9])(00|04|08|12|16|20|24|28|32|36|40|44|48|52|56|60|64|68|72|76|80|84|88|92|96)|(gennaio|febbraio|marzo|aprile|maggio|giugno|luglio|agosto|settembre|ottobre|novembre|dicembre)\\ (19|[2-9][0-9][0-9][0-9])|(\\ ([0-2]*[0-9]|3[0-1])\\ (gennaio|febbraio|marzo|aprile|maggio|giugno|luglio|agosto|settembre|ottobre|novembre|dicembre)\\ (19|[2-9][0-9][0-9][0-9]))|(\\ ([0-2]*[0-9]|3[0-1])\\ (gennaio|febbraio|marzo|aprile|maggio|giugno|luglio|agosto|settembre|ottobre|novembre|dicembre)))))"
    for w in text:
        if "apertura" in w:
            date_apertura="apertura (...)*"
            app=re.search(date_apertura,w)
            if app:
                match=re.search(regex,app.group())
                if match:
                    data=match.group()
                    date.append(data)
        if "inizio" in w:
            date_apertura="inizio (...)*"
            app=re.search(date_apertura,w)
            if app:
                match=re.search(regex,app.group())
                if match:
                    data=match.group()
                    date.append(data)
        if "pubblicato" in w:
            date_apertura="pubblicato (...)*"
            app=re.search(date_apertura,w)
            if app:
                match=re.search(regex,app.group())
                if match:
                    data=match.group()
                    date.append(data)
    return date

def EstrattoreCPV(text):
    regex ="[0-9]{8} [0-9]{1}|[0-9]{8}";
    i_cpv=[]
    for w in text:
        if "cpv" in w:
            r=w.split("cpv")
            match=re.search(regex,r[1])
            if match:
                cpvS=match.group()
                i_cpv.append(cpvS)
    return i_cpv

def EstrattoreCIG(text):
    regex ="\\b(?=.[0-9])[a-z0-9]{10}\\b";
    i_cig=[]
    for w in text:
        if "cig" in w:
            match=list(re.finditer(regex,w))
            for trovato in match:
                cigS=trovato.group()
                i_cig.append(cigS)
    return i_cig

def EstrattoreCUP(text):
    regex ="\\b(?=.?[0-9])[a-z0-9]{3}[a-z0-9]{3}[0-9]{4}[0-9]{5}\\b";
    i_cup=[]
    for w in text:
        if "cup" in w:
            match=list(re.finditer(regex,w))
            for trovato in match:
                cupS=trovato.group()
                i_cup.append(cupS)
    return i_cup

def EstrattoreCategorie(text):
    regex ="\\bog[1-9][0-3]{0,1}|os(1([0-1]|2-a|2-b|[3-7]|8-a|8-b|9)|20-a|20-b|2[1-9]|3[0-5]|(1|2-a|2-b|[3-9]))\\b";
    le_categorie=[]
    for w in text:
        x=w.split()
        for y in x:
            match=re.search(regex,y)
            if match:
                categorie=match.group()
                le_categorie.append(categorie)
    return le_categorie 

def EstrattoreImporti(text):
    gli_importi=[]
    regex="[0-9]{1,3},[0-9]{1,3}|[0-9]{0,3} [0-9]{1,3}.[0-9]{1,3}|[0-9]{0,3} [0-9]{1,3},[0-9]{1,3}|[0-9]{0,3}.[0-9]{0,3}.[0-9]{0,3}.[0-9]{0,3}.[0-9]{2,3},[0-9]{1,3}|[0-9]{0,3}.[0-9]{0,3}.[0-9]{0,3}.[0-9]{0,3}.[0-9]{2,3}.[0-9]{1,3}|[0-9]{0,3} [0-9]{0,3} [0-9]{0,3} [0-9]{0,3} [0-9]{1,3}.[0-9]{1,3}|[0-9]{0,3} [0-9]{0,3} [0-9]{0,3} [0-9]{0,3} [0-9]{1,3},[0-9]{1,3}"
    for w in text:
        if "€" in w:
            match2=list(re.finditer(regex,w))
            for trovato in match2:
                importi=trovato.group()
                gli_importi.append(importi)
        if "euro" in w:
            match2=list(re.finditer(regex,w))
            for trovato in match2:
                importi=trovato.group()
                gli_importi.append(importi)
        if "eur" in w:
            match2=list(re.finditer(regex,w))
            for trovato in match2:
                importi=trovato.group()
                gli_importi.append(importi)
    return gli_importi

def prendoImporto(importi):
    finali=[]
    regex="[0-9]{1,3}.[0-9]{1,3}.[0-9]{1,3}.[0-9]{1,3}|[0-9]{1,3}.[0-9]{1,3}.[0-9]{1,3}|[0-9]{1,3}.[0-9]{1,3}"
    r="[0-9]{1,3}/[0-9]{1,3}"
    r3="[0-9]{0,3}.[0-9]{1,3}.[0-9]{3}"
    for importo in importi:
        match=re.search(regex,importo)
        if match:
            i=match.group()
            i2=i.replace(" ",".",1)
            i3=i2.split(",",1)
            finali.append(i3[0])
    for finale in finali:
        m=re.search(r,finale)
        if m:
            finali.remove(finale) 
    for finale in finali:
        m2=re.search(regex,finale)
        if not m2:
            finali.remove(finale) 
    for finale in finali:
        m3=re.search(r3,finale)
        if not m3:
            finali.remove(finale) 
    for finale in finali:
        if "." not in finale:
            finali.remove(finale) 
    return finali

def pulisciImporti(importi):
    finali=[]
    r3="[0-9]{0,3}.[0-9]{1,3}.[0-9]{3}"
    r2="[0-9].[0-9]"
    punteggiatura=["1.1","1.2","1.3","1.4","1.5","1.6","2.1","2.2","2.3","2.4","2.5","3.1","3.2","3.3","3.4","4.1","4.2","4.3"]

    for i in importi:
        if "."in i:
            finali.append(i)
    for finale in finali:
        m3=re.search(r3,finale)
        if not m3:
            finali.remove(finale) 
    for i in finali:
        if i in punteggiatura:
            finali.remove(i)       
    return finali   

def rimozione_punteggiatura(text):
    words=nltk.sent_tokenize(text)
    frasi=[ ]
    punteggiatura="='':.?:!.,;)/''('\\''//''://''II''III''IV''VI''@'').''),''+''-,''"
    for word in words:
        frasi.append(punt.tokenize(word))
    
    for w in frasi: 
        for word in w:
            if word in punteggiatura :
                w.remove(word)
    return frasi

def riduzione_frasi(frasi):
    f=[] 
    s=""
    for frase in frasi:
            s=" ".join(str(x) for x in frase)
            f.append(s)
    return f

def minuscolo(frasi):
    words=[]
    for w in frasi:
        words.append(w.lower())
    return words   

def elimina_ridondanze(x):
    y=[]
    for w in x:
        if w not in y:
            y.append(str(w))
    return y

def CorrectData(data):
    giusta=[]
    for d in data:
        if d[0]==" ":
            d3=d.replace(" ","",1)
        else:
            d3=d
        d2=ConverterDataNum(d3)
        g=UniformData(d2)
        giusta.append(g)
    return giusta

def ConverterDataNum(data):
    if "gennaio" in data:
        d=data.replace("gennaio","01")
        return d
    elif "febbraio" in data:
        d=data.replace("febbraio","02")
        return d
    elif "marzo" in data:
        d=data.replace("marzo","03")
        return d
    elif "aprile" in data:
        d=data.replace("aprile","04")
        return d
    elif "maggio" in data:
        d=data.replace("maggio","05")
        return d
    elif "giugno" in data:
        d=data.replace("giugno","06")
        return d
    elif "luglio" in data:
        d=data.replace("luglio","07")
        return d
    elif "agosto" in data:
        d=data.replace("agosto","08")
        return d
    elif "settembre" in data:
        d=data.replace("settembre","09")
        return d
    elif "ottobre" in data:
        d=data.replace("ottobre","10")
        return d
    elif "novembre" in data:
        d=data.replace("novembre","11")
        return d
    elif "dicembre" in data:
        d=data.replace("dicembre","12")
        return d
    else:
        return data

def UniformData(data):
    d=data.replace(" ","/")
    return d

def NormalizeCPV(cpvs):
    normalize=[]
    for cpv in cpvs:
        c=cpv.replace(" ","-")
        normalize.append(c)
    return normalize

def ControlloData(date_scadenza,date_apertura):
    final=[]
    for d in date_scadenza:
        if d not in date_apertura:
            final.append(d)
    return final

def rimuovi_punti_importi(importi):
    finali=[]
    for importo in importi:
        i= importo.replace(".","")
        finali.append(i)
    return finali
    
def JSONParse(frasi_lower,frasi_lower2):
    importi2=EstrattoreImporti(frasi_lower2)
    importi=elimina_ridondanze(importi2)
    finali1=prendoImporto(importi)
    finali2=pulisciImporti(finali1)
    finali=rimuovi_punti_importi(finali2)
    finali=elimina_ridondanze(finali)
    cpv2=EstrattoreCPV(frasi_lower)
    cpv3=elimina_ridondanze(cpv2)
    cpv=NormalizeCPV(cpv3)
    cup2=EstrattoreCUP(frasi_lower)
    cup=elimina_ridondanze(cup2)
    cig2=EstrattoreCIG(frasi_lower)
    cig=elimina_ridondanze(cig2)
    nuts2=EstrattoreNUTS(frasi_lower)
    nuts=elimina_ridondanze(nuts2)
    categorie2=EstrattoreCategorie(frasi_lower)
    categorie3=elimina_ridondanze(categorie2)
    categorie=ConvertitoreCategorie(categorie3)
    date_scadenza2=EstrattoreDateScadenza(frasi_lower)
    date_scadenza1=elimina_ridondanze(date_scadenza2)
    date_scadenza3=CorrectData(date_scadenza1)
    date_apertura2=EstrattoreDateApertura(frasi_lower)
    date_apertura1=elimina_ridondanze(date_apertura2)
    date_apertura=CorrectData(date_apertura1)
    date_scadenza=ControlloData(date_scadenza3,date_apertura)
    x = {
    "importi": finali,
    "cpv": cpv,
    "cup": cup,
    "cig": cig,
    "nuts": nuts,
    "categorie":categorie,
    "date_scadenza":date_scadenza,
    "date_apertura":date_apertura
    }
    y=json.dumps(x, ensure_ascii=False)
    return y

def JSONParse2(frasi_lower,frasi_lower2,parole_chiave2):
    importi2=EstrattoreImporti(frasi_lower2)
    importi=elimina_ridondanze(importi2)
    finali1=prendoImporto(importi)
    finali2=pulisciImporti(finali1)
    finali=rimuovi_punti_importi(finali2)
    finali=elimina_ridondanze(finali)
    cpv2=EstrattoreCPV(frasi_lower)
    cpv3=elimina_ridondanze(cpv2)
    cpv=NormalizeCPV(cpv3)
    cup2=EstrattoreCUP(frasi_lower)
    cup=elimina_ridondanze(cup2)
    cig2=EstrattoreCIG(frasi_lower)
    cig=elimina_ridondanze(cig2)
    nuts2=EstrattoreNUTS(frasi_lower)
    nuts=elimina_ridondanze(nuts2)
    categorie2=EstrattoreCategorie(frasi_lower)
    categorie3=elimina_ridondanze(categorie2)
    categorie=ConvertitoreCategorie(categorie3)
    date_scadenza2=EstrattoreDateScadenza(frasi_lower)
    date_scadenza1=elimina_ridondanze(date_scadenza2)
    date_scadenza3=CorrectData(date_scadenza1)
    date_apertura2=EstrattoreDateApertura(frasi_lower)
    date_apertura1=elimina_ridondanze(date_apertura2)
    date_apertura=CorrectData(date_apertura1)
    date_scadenza=ControlloData(date_scadenza3,date_apertura)
    parole_chiave=elimina_ridondanze(parole_chiave2)
    x = {
    "importi": finali,
    "cpv": cpv,
    "cup": cup,
    "cig": cig,
    "nuts": nuts,
    "categorie":categorie,
    "date_scadenza":date_scadenza,
    "date_apertura":date_apertura,
    "parole_chiave":parole_chiave
    }
    y=json.dumps(x, ensure_ascii=False)
    return y

def EstrattoreJSONBandi(bando):
    frasi=rimozione_punteggiatura(bando)
    frasi_stringhe=riduzione_frasi(frasi)
    frasi_lower=minuscolo(frasi_stringhe)
    words=nltk.sent_tokenize(bando)
    frasi_lower2=minuscolo(words)
    b=JSONParse(frasi_lower,frasi_lower2) 
    return b

  
def EstrattoreJSONBandi2(bando,parole_chiave2):
    frasi=rimozione_punteggiatura(bando)
    frasi_stringhe=riduzione_frasi(frasi)
    frasi_lower=minuscolo(frasi_stringhe)
    words=nltk.sent_tokenize(bando)
    frasi_lower2=minuscolo(words)
    b=JSONParse2(frasi_lower,frasi_lower2,parole_chiave2) 
    return b

def stemming(array):
    italianStemmer=SnowballStemmer("italian", ignore_stopwords=True)
    stemm=[]
    for word in array:
        stemm.append(italianStemmer.stem(word))
    return stemm

def sinonimi(parola): #restituisce lista sinonimi singola parola
    synonyms = []
    for syn in wordnet.synsets(parola,lang="ita"):
        for l in syn.lemmas(lang="ita"):
            synonyms.append(l.name())
    return synonyms

def sinonimi_array(parole): #restituisce un unica lista di sinonimi di n parole
    sinonimi2=[]
    for parola in parole:
        p=sinonimi(parola)
        for sin in p:
            sinonimi2.append(sin)
    sinonim=elimina_ridondanze(sinonimi2)
    return sinonim


def parole_chiave(*a):
    sinonim=[]
    for var in a:
        sinonim.append(var)
        app=sinonimi(var)
        for s in app:
            sinonim.append(s)
    sinonim2=elimina_ridondanze(sinonim)
    stemm2=stemming(sinonim2)
    return stemm2


def confronto(stem1,stem2):
    chiavi=[]
    for word in stem1:
        for w in stem2:
            if w in word:
                chiavi.append(w)
    if len(chiavi)==0:
        return False
    else:
        return chiavi
            
def rimozione_punteggiatura_split(testo):
    #divide in parole
    sentence_words = nltk.word_tokenize(testo)
    punteggiatura="='':.?:!.,;)/''('\\''//''://''II''III''IV''VI''@'').''),''+''-,''"
    for word in sentence_words:
        if word in punteggiatura:
            sentence_words.remove(word)
    return sentence_words  

def rimozione_punteggiatura2(text):
    words=nltk.sent_tokenize(text)
    frasi=[ ]
    punteggiatura=["1.1","1.2","1.3","1.4","1.5","1.6","2.1","2.2","2.3","2.4","2.5","3.1","3.2","3.3","3.4","4.1","4.2","4.3"]
    for word in words:
        frasi.append(word.split())
    for w in frasi: 
        for word in w:
            if word in punteggiatura :
                print(word)
                w.remove(word)
    return frasi

def TrovaBando(bando,*a):
    stem_bando=[]
    parole=rimozione_punteggiatura_split(bando)
    parole_lower=minuscolo(parole)
    stem=stemming(parole_lower)
    stem2=parole_chiave(*a)
    result=confronto(stem,stem2) #confronto tra parole chiave e parole nel bando
    if result!= False:
        frasi=rimozione_punteggiatura(bando)
        frasi_stringhe=riduzione_frasi(frasi)
        frasi_lower=minuscolo(frasi_stringhe)
        words=nltk.sent_tokenize(bando)
        frasi_lower2=minuscolo(words)
        b=JSONParse2(frasi_lower,frasi_lower2,result) 
        return b
    else:
        return False

def Trova_Bandi_ParoleChiave(bandi,*a):
    bandi_trovati=[]
    for bando in bandi:
        b=TrovaBando(bando,*a)
        if b!= False:
            bandi_trovati.append(b)
    return bandi_trovati  
        
def ConvertitoreCategorie(category):
    categorie=["og1","og2","og3","og4","og5","og6","og7","og8","og9","og10","og11","og12","og13","os1","os2","os3","os4","os5","os6","os7","os8","os9","os10","os11","os12","os13","os14","os15","os16","os17","os18","os19","os20","os21","os22","os23","os24","os25","os26","os27","os28","os29","os30","os31","os32","os33","os34","os35"]
    finali=[]
    for cat in category:
        if cat in categorie:
            finali.append(categorie.index(cat)+1)
    return finali
        
def AccessJsonToCSV(jsoon,pkid):
    j=json.loads(jsoon)
    cpv=transformArray(j["cpv"])
    cup=transformArray(j["cup"])
    cig=transformArray(j["cig"])
    importi=transformArray(j["importi"])
    categorie=transformArray(j["categorie"])
    nuts=transformArray(j["nuts"])
    date_scadenza=transformArray(j["date_scadenza"])
    date_apertura=transformArray(j["date_apertura"])
    bandi_writer.writerow([pkid,cpv,cup,cig,importi,categorie,nuts,date_scadenza,date_apertura])
    
def transformArray(array):
    s=""
    if len(array)==1:
        for a in array:
            s=a
            return s
    elif len(array)==0:
        return s
    else:
        for a in array:
            s=s+str(a)+","
    b=s.rstrip(",")
    return b

 
print("OUTPUT ESTRATTORE")
print("-----------------------------------------------------------------------------------")
print("Outpu bando 1: ")
json_bandi1=EstrattoreJSONBandi(bando1)
print(json_bandi1)
print("-----------------------------------------------------------------------------------")
print("Outpu bando 2: ")
json_bandi2=EstrattoreJSONBandi(bando2)
print(json_bandi2)
#stampa_controllo(json_bandi)
#synonyms=sinonimi("piatto")
parole=['finestra','infissi']
out=stemming(parole)
print(out)
print("-----------------------------------------------------------------------------------")
parola=input("Inserire parole chiave che si desidera ricercare nei bandi: ")
print("-----------------------------------------------------------------------------------")
print("Stampa dei bandi che contengono la parola chiave: "+parola)
Trova_Bandi_ParoleChiave(bandi,parola)
#stampa_controllo(frasi_lower)
```
